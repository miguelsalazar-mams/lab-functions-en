## 2. Now write a function that returns the largest element on a list

```python
#your code here
def greatest(my_list):
    largest = my_list[0]
    for element in my_list:
        if element > largest:
            largest = element
    return largest
```

## 7. Write a function that takes a list and returns a list of the unique values.

`NOTE: You cannot use set. 🤔`

```python
#STEP 1: CREATE A NEW EMPTY LIST TO STOR UNIQUE VALUES
#STEP 2: LOOP THROUGH EACH ITEM IN THE INPUT LIST
#STEP 3: IF THE ITEM IS NOT ALREADY IN THE NEW LIST, THEN ADD IT (APPEND)
#STEP 4: AFTER THE LOOP, RETURN THE NEW LIST


def unique(lst_un):
    uniques = []
    for item in lst_un:
        if item not in uniques:
            uniques.append(item)
    return uniques
```
